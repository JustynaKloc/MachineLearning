<H3>Ćwiczenie 1</H3>

Celem ćwiczenia jest:
- zapoznanie z narzędziami do wczytywania zbiorów zawierających przykłady w postaci wektorów cech oraz nadanie danym uporządkowanej reprezentacji;

- ekstrakcja wybranych informacji z utworzonych struktur danych.


___

Import potrzebnych bibliotek:
 - sklearn (scikit-learn) to pakiet wspomagający uczenie maszynowe.
 - matplotlib umożliwia tworzenie wykrecsów i wizualizację danych.

```python
import sklearn as sk
import matplotlib.pyplot as plt
```

Pobranie bazy danych z zasobów sklearn.

```python
from sklearn import datasets
iris = datasets.load_iris()
X_iris, y_iris = iris.data, iris.target #podział danych na macierz cech i wektor klas
print (X_iris.shape, y_iris.shape)
print (X_iris[0], y_iris[0])
print (iris.target_names)   #nazwy klas
print (iris.feature_names)  #nazwy cech
#print (iris.DESCR)
```

Wybieranie danych z macierzy

```python
X_iris[:10]#pierwsze 10 wartości
```

```python
X_iris[3:9:2]#wartości od 3 do 9 co druga, bez ostatniej wartosci - przedział <3,9)
```

```python
X_iris[:10,2]#wybór drugiej kolumny (numerowanie od 0) z pierwszych 10 wartości
```

```python
X_iris[ X_iris[:,1]>3.5 ]#wybór obiektów, których wartość cechy w pierwszej kolumnie (sepal width (cm)) jest większa od 3.5
```

Wyświetlanie zbioru danych w 2D

```python
plt.figure()
featurex=2
featurey=1
plt.scatter(X_iris[y_iris==0,featurex],X_iris[y_iris==0,featurey], c='b', marker='x', label=iris.target_names[0])
plt.scatter(X_iris[y_iris==1,featurex],X_iris[y_iris==1,featurey], c='g', marker='o', label=iris.target_names[1])
plt.scatter(X_iris[y_iris==2,featurex],X_iris[y_iris==2,featurey], c='r', marker='^', label=iris.target_names[2])
plt.legend(loc='upper right')
plt.axis('equal')
plt.show()
```

Normalizowanie zbioru danych

```python
iris_normalized=(X_iris-np.tile(np.min(X_iris,axis=0), (150,1))) / (np.tile(np.max(X_iris,axis=0)-np.min(X_iris,axis=0), (150,1)))
```

```python
plt.figure()
featurex=2
featurey=1
plt.scatter(iris_normalized[y_iris==0,featurex],iris_normalized[y_iris==0,featurey], c='b', marker='x', label=iris.target_names[0])
plt.scatter(iris_normalized[y_iris==1,featurex],iris_normalized[y_iris==1,featurey], c='g', marker='o', label=iris.target_names[1])
plt.scatter(iris_normalized[y_iris==2,featurex],iris_normalized[y_iris==2,featurey], c='r', marker='^', label=iris.target_names[2])
plt.legend(loc='upper right')
plt.axis('equal')
plt.show()
```

Wyświetlanie zbioru danych w 3D

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_iris[y_iris==0,0],X_iris[y_iris==0,1],X_iris[y_iris==0,2], c='b', marker='x', label=iris.target_names[0])
ax.scatter(X_iris[y_iris==1,0],X_iris[y_iris==1,1],X_iris[y_iris==1,2], c='g', marker='o', label=iris.target_names[1])
ax.scatter(X_iris[y_iris==2,0],X_iris[y_iris==2,1],X_iris[y_iris==2,2], c='r', marker='^', label=iris.target_names[2])

ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(iris_normalized[y_iris==0,0],iris_normalized[y_iris==0,1],iris_normalized[y_iris==0,2], c='b', marker='x', label=iris.target_names[0])
ax.scatter(iris_normalized[y_iris==1,0],iris_normalized[y_iris==1,1],iris_normalized[y_iris==1,2], c='g', marker='o', label=iris.target_names[1])
ax.scatter(iris_normalized[y_iris==2,0],iris_normalized[y_iris==2,1],iris_normalized[y_iris==2,2], c='r', marker='^', label=iris.target_names[2])

ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])

plt.show()
```

Zmiana sposobu wyświetlania wykresów

```python
#%matplotlib inline 
%matplotlib qt5
```

Wizualizacja danych wielowymiarowych na płaszczyźnie

```python
from sklearn.manifold import TSNE

model = TSNE(learning_rate=100)
transformed = model.fit_transform(X_iris)

xs=transformed[:,0]
ys=transformed[:,1]

plt.figure()
plt.scatter(xs,ys,c=y_iris)
plt.show()
```
