```python
#plot_surface do narysowania płaszczyny pozwalającej na rozróżnienie dwóch klas 
from sklearn.svm import SVC

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

svc = SVC(kernel='linear')
svc.fit(X_train,y_data)

z = lambda x,y: (-svc.intercept_[0]-svc.coef_[0][0]*x-svc.coef_[0][1]*y) / svc.coef_[0][2]
tmp = np.linspace(-300,2000,51)
x,y = np.meshgrid(tmp,tmp)

ax.plot_surface(x,y,z(x,y))

ax.scatter(X_train[y_data==0,0],X_train[y_data==0,1],X_train[y_data==0,2], c='b', marker='x', label=dataset.target_names[0])
ax.scatter(X_train[y_data==1,0],X_train[y_data==1,1],X_train[y_data==1,2], c='g', marker='o', label=dataset.target_names[1])

ax.set_xlabel(dataset.feature_names[0])
ax.set_ylabel(dataset.feature_names[1])

plt.show()
```

Korzystając z pliku csv wczytaj obrazy z bazy danych. <br>
Następnie wyświetl 25 losowych obrazów na jednej kanwie, np. używając subplots.

```python
import csv
database = []

path = 'GTSRB/Final_Training/Images/00000/'
with open('GTSRB/Final_Training/Images/00000/GT-00000.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ')
    line_count = 0
    for row in spamreader:
        if line_count == 0:
            line_count +=1
        else:
            #print(row[0])
            row =  ', '.join(row)
            row = row.split(';')
            save = row[0]          
            database.append(imread(path + save))
#print(database)
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.grid(False)
    plt.imshow(database[i])
plt.show()
```
